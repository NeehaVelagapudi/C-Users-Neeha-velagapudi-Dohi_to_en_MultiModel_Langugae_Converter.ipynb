```python
import torch
from transformers import MarianMTModel, MarianTokenizer, pipeline
import vosk
import json
import wave
import librosa
import IPython.display as ipd
from google.colab import files
from TTS.api import TTS
```

```python
# Load English-to-Hindi translation model
translation_model_name = "Helsinki-NLP/opus-mt-hi-en"
translation_tokenizer = MarianTokenizer.from_pretrained(translation_model_name)
translation_model = MarianMTModel.from_pretrained(translation_model_name)
!pip install sacremoses
```

```python
# Specify the target language
target_language = "en"  # english
!apt-get install espeak-ng  # Install espeak-ng dependency
from TTS.api import TTS
tts_model = TTS("tts_models/multilingual/multi-dataset/your_tts")
# Load Vosk speech recognition model
vosk_model = vosk.Model("vosk-model-hi-0.22")
```

```python
def transcribe_speech(filename):
    """Converts speech to text using Vosk ASR."""
    wf = wave.open(filename, "rb")
    rec = vosk.KaldiRecognizer(vosk_model, wf.getframerate())
    text = ""
    while True:
        data = wf.readframes(4000)
        if len(data) == 0:
            break
        if rec.AcceptWaveform(data):
            result = json.loads(rec.Result())
            text += result.get("text", "") + " "
    return text.strip()
```

```python
"""Translates English text to Hindi."""
def generate_translation(input_text):
  inputs = translation_tokenizer(input_text, return_tensors="pt", padding=True, truncation=True)
  translated_tokens = translation_model.generate(**inputs)
  translated_text = translation_tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0]
  return translated_text
```

```python
def synthesize_speech(text):
    """Converts translated text to speech using an available language."""
    audio_path = "translated_speech.wav"
    available_speakers = tts_model.speakers
    print("Available speakers:", available_speakers)
    speaker_name = available_speakers[0]  # Choose the first available speaker
    language_name = "en"  # Hindi ('hi') is not supported, so use English ('en')
    tts_model.tts_to_file(
        text=text,
        file_path=audio_path,
        speaker=speaker_name,
        language=language_name )
    return audio_path
```

फाइव आइज अलायंस क्या है? इसकी शुरुआत कैसे हुई? अमेरिका में कनाडा को इस गठबंधन से निकालने की चर्चा क्यों शुरू हुई और यह कहां तक पहुंची है? फाइल आइज अलायंस काम कैसे करता है और कनाडा की खुफिया क्षमताओं में इसकी क्या भूमिका है? कनाडा में खालिस्तानी आतंकी हरदीप सिंह निज्जर की हत्या में इस गठबंधन से कनाडा को किस तरह की मदद मिली थी? आइये जानते हैं|
अमेरिका में डोनाल्ड ट्रंप की सरकार बनने के बाद से ही कुछ देशों में हलचल मची है। इनमें सबसे अहम नाम अमेरिका के मित्र और सहयोगी देशों के हैं। हालांकि, जिस एक देश को ट्रंप ने सबसे ज्यादा निशाना बनाया है, वह अमेरिका के उत्तर में उसका सबसे करीबी पड़ोसी देश- कनाडा। चाहे आयात शुल्क लगाने से जुड़ी धमकियां हों या कनाडा से लगी सीमाओं पर अतिरिक्त सख्ती बरतने की चेतावनी या फिर कनाडा को अमेरिका का 51वां राज्य बनाने से जुड़ा तंज। ट्रंप हर एक मौके पर कनाडा पर हावी होने की कोशिश में रहे हैं।
हालांकि, जिस ताजा मामले पर अमेरिका की तरफ से कनाडा के लिए सख्त रुख अख्तियार करने का इशारा किया गया है, वह उसकी सुरक्षा को सीधे तौर पर प्रभावित करता है। मामला है कनाडा को एक ऐसे फाइव आइज अलायंस से बाहर करने की चर्चा का। अगर कनाडा आगामी समय में इस गठबंधन से बाहर होता है, तो इसके कई दूरगामी असर हो सकते हैं। खासकर कनाडा की खुफिया और निगरानी क्षमताओं पर नकारात्मक असर पड़ना तय है।

```python
def main():
    """Main function for speech/text-based Hindi-to-English translation."""
    user_choice = input("Enter 'text' for text input or 'speech' for audio input: ")

    if user_choice == 'text':
        input_text = input("Enter your text in Hindi: ")
    elif user_choice == 'speech':
        uploaded = files.upload()
        if not uploaded:
            print("No file uploaded.")
            return
        filename = next(iter(uploaded))
        input_text = transcribe_speech(filename)
        print(f"Transcribed Text: {input_text}")
    else:
        print("Invalid input type. Please enter 'text' or 'speech'.")
        return

    # Generate translation
    translated_text = generate_translation(input_text)
    print("Translated Text:", translated_text)

    # Convert translated text to speech
    audio_path = synthesize_speech(translated_text)
    ipd.display(ipd.Audio(audio_path))

# Run the program
main()
```
